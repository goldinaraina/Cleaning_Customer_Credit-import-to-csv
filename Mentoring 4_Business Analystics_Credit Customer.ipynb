# ``DATA CLEANING CREDIT CUSTOMER``

```python
import csv

# menampilkan dataset
df1 = pd.read_csv('application_record.csv')

df1.head()
```

## Missing value dan validasi

```python
# melakukan drop pada baris yang memiliki missing value
df1 = df1.dropna(subset=['OCCUPATION_TYPE'])

df1.shape
```

```python
# memvalidasi dataset df1
df1.isnull().sum()
```

```python
# menampilkan dataset kedua
df2 = pd.read_csv('credit_record.csv')

df2.head()
```

## Gabungkan Dataset Based on ID

```python
merged_df = pd.merge(df1, df2, on='ID')

merged_df.head(5)
```

```python
# memvalidasi dataset
merged_df.isnull().sum()
```

## Cek Deskritif Dataset

```python
merged_df.describe()
```

## Merubah Kolom DAYS_BIRTH dan DAYS_EMPLOYED 

```python
# hitung usia dalam tahun
merged_df['AGE_YEARS'] = merged_df['DAYS_BIRTH'].apply(lambda x: abs(x) // 365)

# hitung lama bekerja dalam tahun 
merged_df['YEARS_EMPLOYED'] = merged_df['DAYS_EMPLOYED'].apply(lambda x: x // 365 if x < 0 else 0)

# menampilkan output
merged_df[['DAYS_BIRTH', 
           'AGE_YEARS', 
           'DAYS_EMPLOYED', 
           'YEARS_EMPLOYED']]
```
